```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/car price prediction.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
df = data['Market Category'].str.get_dummies(sep = ',')
df.head()
```

```python
data = pd.concat([data, df], axis = 1)
```

```python
data.columns
```

```python
data.head()
```

```python
data.drop(columns = 'Market Category', inplace = True)
```

```python
X = data.drop(columns = 'MSRP')
y = data['MSRP']
```

```python
X.head()
```

```python
X.info()
```

```python
category = []

for column in X.columns:
  if X[column].dtypes == 'O':
    category.append(column)

integers = []

for column in X.columns:
  if X[column].dtypes == 'int64':
    integers.append(column)


floats = []

for column in X.columns:
  if X[column].dtypes == 'float64':
    floats.append(column)
```

```python
for column in floats:
  X[column] = X[column].astype('int64')
```

```python
X.info()
```

```python
X[category].head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for column in category:
  X[column] = le.fit_transform(X[column])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

print(lr.intercept_)
print(lr.coef_)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE: ', mse)
print('MAE: ', mae)
print('R2: ', r2)
```

```python
from sklearn.svm import SVR

svr = SVR(kernel = 'linear')
svr.fit(X_train, y_train)
```

```python
print("y = {0} + {1}*x".format(svr.intercept_[0], svr.coef_[0][0]))
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('MSE: ', mse)
print('RMSE: ', rmse)
print('MAE: ', mae)
print('R2: ', r2)
```
