```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/car price prediction.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
df = data['Market Category'].str.get_dummies(sep = ',')
```

```python
df.head()
```

```python
data = pd.concat([data, df], axis = 1)
```

```python
data.columns
```

```python
data.head()
```

```python
data.drop(columns = 'Market Category', inplace = True)
```

```python
X = data.drop(columns = 'MSRP')
y = data['MSRP']
```

```python
X.head()
```

```python
X.info()
```

```python
category = []

for column in X.columns:
  if X[column].dtypes == 'O':
    category.append(column)

integers = []

for column in X.columns:
  if X[column].dtypes == 'int64':
    integers.append(column)


floats = []

for column in X.columns:
  if X[column].dtypes == 'float64':
    floats.append(column)
```

```python
X['Number of Doors'].dtypes
```

```python
floats
```

```python
X[floats].head()
```

```python
X[integers].head()
```

```python
for column in floats:
  X[column] = X[column].astype('int64')
```

```python
X.info()
```

```python
X[category].head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for column in category:
  X[column] = le.fit_transform(X[column])
```

```python
X.head()
```

```python
X['Make'].unique()
```

```python
X['Year'].unique()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
dtr = DecisionTreeRegressor()

dtr.fit(X_train, y_train)
```

```python
y_pred = dtr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def metrics(y_test, y_pred):
  mse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  rmse = np.sqrt(mse)
  r2 = r2_score(y_test, y_pred)

  print('MSE = ', mse)
  print('RMSE = ', rmse)
  print('MAE = ', mae)
  print('r2 = ', r2)
```

```python
metrics(y_test, y_pred)
```

```python
rf = RandomForestRegressor()

rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
metrics(y_test, y_pred)
```
