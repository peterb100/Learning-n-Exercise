```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/xAPI-Edu-Data.xls')
```

```python
data.head()
```

```python
data.info()
```

```python
data['Class'].unique()
```

```python
from pandas.core.resample import date_range
fig, ax = plt.subplots(2, 2, figsize = (10, 10))

sns.countplot(x = 'Class', data = data, ax = ax[0,0], order = ['L', 'M', 'H'])
sns.countplot(x = 'gender', data = data, ax = ax[0,1], order = ['M', 'F'])
sns.countplot(x = 'StageID', data = data, ax = ax[1, 0])
sns.countplot(x = 'Semester', data = data, ax = ax[1, 1])

plt.show()
```

```python
data.columns
```

```python
data = data.rename(columns = {'NationalITy':'Nationality'})
```

```python
fig, (ax1,  ax2) = plt.subplots(2, 1, figsize = (10, 10))

sns.countplot(x = 'Topic', data = data, ax = ax1)
sns.countplot(x = 'Nationality', data = data, ax = ax2)

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize = (10, 10))
sns.countplot(x = 'gender', hue = 'Class', data = data, ax = ax[0, 0])
sns.countplot(x = 'gender', hue = 'Relation', data = data, ax = ax[0, 1])
sns.countplot(x = 'gender', hue = 'StudentAbsenceDays', data = data, ax = ax[1, 0])
sns.countplot(x = 'gender', hue = 'ParentAnsweringSurvey', data = data, ax = ax[1, 1])

plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 10))
sns.countplot(x = 'Topic', hue = 'gender', data = data, ax = ax[0])
sns.countplot(x = 'Nationality', hue = 'gender', data = data, ax = ax[1])
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 10))

sns.countplot(x = 'Nationality', hue = 'Relation', data = data, ax = ax[0])
sns.countplot(x = 'Nationality', hue = 'StudentAbsenceDays', data = data, ax = ax[1])

plt.show()
```

```python
from pandas.core.resample import date_range
fig, ax = plt.subplots(2, 2, figsize = (10, 10))
plt.grid(axis = 'y')

sns.barplot(x = 'Class', y = 'VisITedResources', data = data, ax = ax[0,0])
sns.barplot(x = 'Class', y = 'AnnouncementsView',data = data, ax = ax[0,1])
sns.barplot(x = 'Class', y = 'raisedhands',data = data, ax = ax[1, 0])
sns.barplot(x = 'Class', y = 'Discussion',data = data, ax = ax[1, 1])

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

sns.swarmplot(x = 'gender', y = 'AnnouncementsView', data = data, ax = ax[0])
sns.swarmplot(x = 'gender', y = 'raisedhands', data = data, ax = ax[1])

plt.show()
```

```python
figfig, ax = plt.subplots(1, 2, figsize=(10, 5))

sns.boxplot(x = 'Class', y = 'Discussion', data = data, ax = ax[0])
sns.boxplot(x = 'Class', y = 'VisITedResources', data = data, ax = ax[1])

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))

sns.pointplot(x = 'Semester', y = 'VisITedResources', hue = 'gender', data = data, ax = ax[0])
sns.pointplot(x = 'Semester', y = 'AnnouncementsView', hue = 'gender', data = data, ax = ax[1])

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))

sns.regplot(x = 'raisedhands', y = 'VisITedResources', data = data, ax = ax[0])
sns.regplot(x = 'AnnouncementsView', y = 'Discussion', data = data, ax = ax[1])
```
