```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_rows', 100)
```

```python
b_cancer=pd.read_csv('/content/breast-cancer-data.csv')
```

```python
b_cancer.head()
```

```python
b_cancer.info()
```

```python
b_cancer['age'].unique()
```

```python
fig, (ax1, ax2) = plt.subplots(2,1, figsize = (10,10))
sns.countplot(x='age', data = b_cancer, ax = ax1)
sns.countplot(x='class',data = b_cancer, ax = ax2)
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize = (10,10))
sns.countplot(x ='menopause', hue = 'node-caps',data=b_cancer, ax=ax[0] )
sns.countplot(x ='menopause', hue = 'inv-nodes', data = b_cancer, ax = ax[1] )
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize = (10,10))
sns.countplot(x ='age', hue = 'node-caps',data=b_cancer, ax=ax[0] )
sns.countplot(x ='age', hue = 'inv-nodes', data = b_cancer, ax = ax[1] )
plt.show()
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,10))
sns.countplot(x='breast', data = b_cancer, ax=ax[0])
sns.countplot(x='breast-quad', data = b_cancer, ax=ax[1])
plt.show()
```
