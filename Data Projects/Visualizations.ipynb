```python
def plot_timeseries(axis, x, y, color, xlabel, ylabel):
  axis.plot(x, y, color = color)
  axis.set_xlabel(xlabel)
  axis.set_ylabel(ylabel)
  axis.tick_params('y', colors = color)
```

```python
fig, ax = plt.subplots()

plot_timeseries(ax, merged.index, merged['Interpolated'], 'b', 'Time', 'CO2')

ax2 = ax.twinx()

plot_timeseries(ax2, merged.index, merged['avg_temp_c'], 'r', 'Time', 'Temperature')
```

```python
population = pd.read_csv('https://pkgstore.datahub.io/core/population-city/unsd-citypopulation-year-fm_csv/data/8b8d4d72dbf1a503bc3e241b394ce301/unsd-citypopulation-year-fm_csv.csv')
```

```python
population.head()
```

```python
france = population[(population['Country or Area'] == 'France')&(population['Year'] == '2010')&(population['City type'] == 'City proper')].sort_values(by = ['City']).set_index('City')
```

```python
france.head()
```

```python
france.info()
```

```python
france['Value'] = france['Value'].astype(float).astype(int)
```

```python
population[['Country or Area', 'City']].groupby('Country or Area').count()
```

```python
groupby = france[:60].groupby('Sex')

female, male = [groupby.get_group(x) for x in groupby.groups]
```

```python
fig, ax = plt.subplots()

ax.bar(female.index, female['Value'])
ax.xaxis.set_tick_params(rotation = 80)
ax.set_ylabel('Female population')

ax.grid(axis = 'y')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(female.index, female['Value'], label = 'Female')
ax.bar(male.index, male['Value'], label = 'Male')

ax.xaxis.set_tick_params(rotation = 80)
ax.set_ylabel('pop by sex')

ax.legend()
ax.grid(axis = 'y')
plt.show()
```

```python
width = 0.5

fig, ax = plt.subplots()

ax.bar(female.index, female['Value'], label = 'Female')
ax.bar(male.index, male['Value'], width = width, align = 'edge', label = 'Male')

ax.xaxis.set_tick_params(rotation = 80)
ax.set_ylabel('pop by sex')

ax.legend()
ax.grid(axis = 'y')
plt.show()
```

```python
medals = pd.read_csv('https://raw.githubusercontent.com/matzim95/ML-datasets/master/olympic.csv', index_col = 'Year')
```

```python
medals.head()
```

```python
def groupby_column(dataframe, column):
  groupby = dataframe.groupby(column)

  return {x:groupby.get_group(x) for x in groupby.groups}
```

```python
all_sports = groupby_column(medals, 'Sport')
```

```python
fig, ax = plt.subplots()

ax.bar('Gymnastics', all_sports['Gymnastics']['Height'].mean())
ax.bar('Basketball', all_sports['Basketball']['Height'].mean())

ax.set_ylabel('Height [cm]')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(all_sports['Gymnastics']['Height'], label = 'Gymnastics', bins = 20)
ax.hist(all_sports['Basketball']['Height'], label = 'Basketball', bins = 20, histtype = 'step')

ax.set_xlabel('Height')
ax.set_ylabel('Observations')

ax.legend()
plt.show()
```
